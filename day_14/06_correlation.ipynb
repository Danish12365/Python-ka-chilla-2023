**Correlation**

What is Correlation?\
Variables within a dataset can be related for lots of reasons.

**Types**:

Pearson's r\
Spearman's rho\
Kendall's tau

**For example:**

1.  One variable could cause or depend on the values of another variable.
2.  One variable could be lightly associated with another variable.
3.  Two variables could depend on a third unknown variable.

**Positive Correlation:** both variables change in the same direction.\
**Negative Correlation:** No Relationship in the change of the variables.\
**Negative Correlation:** variables change in opposite direction.

**Covariance**

* Variables can be related by a linear relationship. This is a relationship that is consistently additive across the two data samples.
* This relationship can be summarized between two variables, called the covariance.
* The sign of the covariance can be interpreted as whether the two variables change in the same direction (positive) or change in different directions (negative).
* The magnitude of the covariance is not easily interpreted. A covariance value of zero indicates that both variables are completely independent.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define some data with different correlation strengths
x = np.linspace(0, 10, 100)

# Positive correlation
y1 = 2 * x + 3

# Negative correlation
y2 = -1 * x + 5

# Zero correlation
y3 = np.random.rand(100)

# Slightly positive correlation
y4 = x + np.random.randn(100) * 0.2

# Create the plots
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Plot 1: Positive correlation
axes[0, 0].scatter(x, y1)
axes[0, 0].set_title("Positive Correlation")
axes[0, 0].set_xlabel("X")
axes[0, 0].set_ylabel("Y1")

# Plot 2: Negative correlation
axes[0, 1].scatter(x, y2)
axes[0, 1].set_title("Negative Correlation")
axes[0, 1].set_xlabel("X")
axes[0, 1].set_ylabel("Y2")

# Plot 3: Zero correlation
axes[1, 0].scatter(x, y3)
axes[1, 0].set_title("Zero Correlation")
axes[1, 0].set_xlabel("X")
axes[1, 0].set_ylabel("Y3")

# Plot 4: Slightly positive correlation
axes[1, 1].scatter(x, y4)
axes[1, 1].set_title("Slightly Positive Correlation")
axes[1, 1].set_xlabel("X")
axes[1, 1].set_ylabel("Y4")

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate random data for the four scenarios
size = 100
x = np.random.rand(size)
# Positive correlation
y_pos = 2 * x + 1 + np.random.normal(0, 0.2, size)
# Negative correlation
y_neg = -2 * x + 1 + np.random.normal(0, 0.2, size)
# Zero correlation
y_zero = np.random.normal(0, 1, size)
# Slightly positive/negative correlation
y_slight = x + np.random.normal(0, 0.2, size)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Positive correlation
axs[0, 0].scatter(x, y_pos)
axs[0, 0].set_title('Positive Correlation')

# Negative correlation
axs[0, 1].scatter(x, y_neg)
axs[0, 1].set_title('Negative Correlation')

# Zero correlation
axs[1, 0].scatter(x, y_zero)
axs[1, 0].set_title('Zero Correlation')

# Slightly positive/negative correlation
axs[1, 1].scatter(x, y_slight)
axs[1, 1].set_title('Slightly Positive/Negative Correlation')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
