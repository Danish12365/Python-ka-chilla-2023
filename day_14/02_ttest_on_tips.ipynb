```python
# one sample t-test

# import libraries
import pandas as pd
import seaborn as sns
from  scipy.stats import ttest_1samp

# load dataset

df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df1 = df[['total_bill','tip', 'sex', 'size']]
df1.head()
```

```python
df1.describe()
```

```python
ttest_1samp(df1['size'], 50)
stat, p = ttest_1samp(df1['size'], 50)
print('stat=%.3f, p=%.3f' % (stat, p))

# make a coditional argument for further use
if p > 0.05:
	print('Probably the same Distribution')
else:
	print('Probably different distribution')
```

## Two sample t-tests
Independent student's t-test

***Assumptions***
- Observations in each sample are independent and identically distributed.
- Observations in each sample are normally distributed.
- Observations in each sample have the same variance.
 
***Interpretation***
  
*H0:* the means of the samples are equal.\
*H1:* the means of the samples are unequal.



```python
# we will compare age and fare of male vs female passengers

#splitting datasets
df_male = df1.loc[df1['sex']=='Male']
df_female = df1.loc[df1['sex']=='Female']
#library
from scipy.stats import ttest_ind
stat, p = ttest_ind(df_male['size'], df_female['size'])
print('stat=%.3f, p=%.3f' % (stat, p))

# make a coditional argument for further use
if p > 0.05:
	print('Probably the same Distribution')
else:
	print('Probably different distribution')
```

***Paired student's t-test***

Tests whether the means of two paired samples are significantly different.

***Assumptions***
- Observations in each sample are independent and identically distributed.
- Observations in each sample are normally distributed.
- Observations in each sample have the same variance.
- Observations across each sample are paired.
  
***Interpretation***

*H0:* the means of the samples are equal.\
*H1:* the means of the samples are unequal.

```python
df.head()
```

```python
# select only male's data
df_male = df.loc[df['sex']=='Male']
df_male.head()
```

```python
# select only two classes
df_male_first = df_male.loc[df_male['size']==2]
df_male_second = df_male.loc[df_male['size']==3]
df_male_third = df_male.loc[df_male['size']==4]
```

```python
# df_2= df_male_first.sample(n=100)
# df_3 = df_male_second.sample(n=100)

# print("The number of instances in 1st size are = ", df_2.size)
# print("The number of instances in 2nd size are = ", df_3.size)


# Assuming df_male_first and df_male_second are your original DataFrames

# Check the size of the original DataFrames
size_first = df_male_first.shape[0]
size_second = df_male_second.shape[0]

# Sample with replacement if needed
df_2 = df_male_first.sample(n=min(100, size_first), replace=True)
df_3 = df_male_second.sample(n=min(100, size_second), replace=True)

print("The number of instances in the 1st sample is =", df_2.shape[0])
print("The number of instances in the 2nd sample is =", df_3.shape[0])
```

```python
# import library

# from scipy.stats import ttest_rel
# # apply test to copare class-1 and class-2
# stat, p = ttest_rel(df_2['size'],df_3['size'])
# print('stat=%.3f, p=%.3f' % (stat, p))

# # adding  coditional argument for ease
# if p > 0.05:
#     print('Probably the same Distribution')
# else:
#     print('Probably different distribution')


from scipy.stats import ttest_ind

# Perform independent t-test
stat, p = ttest_ind(df_2['size'], df_3['size'], equal_var=False)  # Use equal_var=False if variances are not assumed to be equal

print('stat=%.3f, p=%.3f' % (stat, p))

# Adding conditional argument for ease
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: There is a significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the groups.")

    
```
