```python
# sns.lmplot(data=df, x='fare', y='age')
# sns.lmplot(data=df, x='age', y=df['fare'] <300)
# sns.lmplot(data=df[df['fare'] < 300], x='age', y='fare )  # assighnment 

# Add a new column based on a condition (fare < 300)
# df['condition'] = df['fare'] < 300

# Create a new column 'condition' based on the fare
df['condition'] = df['fare'].apply(lambda x: 'low' if x < 300 else 'high')

# Use lmplot with hue for conditional plotting
sns.lmplot(data=df, x='age', y='fare', hue='condition', markers=["o", "s"], palette={"low": "blue", "high": "red"})
```

```python

# familey_size
# df['familey_size'] = pd.cut([])

# Define a function to determine family size
def get_family_size(row):
    if row['sibsp'] == 0 and row['parch'] == 0:
        return 'Single'
    elif row['sibsp'] == 1 and row['parch'] == 0:
        return 'Couple (No Children)'
    elif row['sibsp'] > 1 and row['parch'] == 0:
        return 'Family (No Children)'
    elif row['sibsp'] == 0 and row['parch'] == 1:
        return 'Single Parent'
    elif row['sibsp'] == 1 and row['parch'] == 1:
        return 'Nuclear Family'
    elif row['sibsp'] > 1 and row['parch'] == 1:
        return 'Extended Family'

# Create 'family_size' column based on conditions
df['family_size'] = df.apply(get_family_size, axis=1)
```

```python
df.columns
```

```python
df.groupby("family_size")["fare"].mean().plot(kind="bar")
```
