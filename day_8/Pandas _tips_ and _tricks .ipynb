# 16- Splitting a string into multiple coulumns

```python
# import libraries
import pandas as pd
df = pd.DataFrame({'Name':['Ahmad Raza', 'Ali Afzal', 'Sajjad Ali', 'Abu Bakkar'],
                   'location':['Lahore, pakistan','Sargodah, Pakistan', 'Karachi, Pakistan','Hamburg, Germany']})
df
                   
```

```python
# spliting a columns into two columns
df.Name.str.split(' ', expand=True)
```

```python
# addding those splits into new columns
df[["first_Name", "last_Name"]] = df.Name.str.split(' ', expand=True)
df
```

```python
df[["city", "country"]] = df.location.str.split(', ', expand=True)
df
```

```python
# Refine data manipulation
df = df[['first_Name', 'last_Name', 'city', 'country']]
df
```
