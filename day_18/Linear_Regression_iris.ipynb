```python
x = multilinear_regression[['sepal_length', 'sepal_width','petal_length']]
y = multilinear_regression['petal_width']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# model bula len
model = LinearRegression()

# train the model
model.fit(x,y)

#prediction
model.predict([[1.5,2.5,3.5]])
```

```python
#model evalution with metrics
# r2
# mean squared error
# MAE
# RMSE

x_train, x_test, y_train, y_test = train_test_split(x,y, train_size= 0.8)

# model
model = LinearRegression()

# train
model.fit(x_train, y_train)

# prediction
predictions = model.predict(x_test)
```

```python
sns.scatterplot(x= y_test, y= predictions)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

MAE = mean_absolute_error(y_true=y_test, y_pred=predictions)
MSE = mean_squared_error(y_true=y_test, y_pred=predictions)
RMSE = mean_squared_error(y_true=y_test, y_pred=predictions, squared= False)
R2 = r2_score(y_true=y_test, y_pred=predictions)

print("MAE: ", MAE)
print("MSE: ",MSE)
print("RMSE: ",RMSE)
print("R_squared: ",R2)
```

**Here's an assessment of the model's performance based on the provided metrics:**

**1. MAE (Mean Absolute Error):**
   - Value: 0.16636417029533254
   - Interpretation: On average, the model's predictions are off by about 0.166 units from the actual values. This is relatively low, suggesting good accuracy.

**2. MSE (Mean Squared Error):**
   - Value: 0.05312437165807597
   - Interpretation: This measures the average squared difference between predictions and actual values. It's lower than the MAE, indicating a good fit.

**3. RMSE (Root Mean Squared Error):**
   - Value: 0.23048724836327925
   - Interpretation: This is in the same units as the target variable and gives an idea of the average magnitude of the errors. It's relatively low in this case.

**4. R_squared (R-squared):**
   - Value: 0.8975860886960086
   - Interpretation: This is quite high, indicating that the model explains 89.76% of the variance in the target variable. This is a strong result.

**Overall Assessment:**

- Based on these metrics, the model is performing well. It has a low MAE, low MSE, low RMSE, and a high R-squared value. These scores suggest that it's accurately capturing the patterns in the data and making reliable predictions.
